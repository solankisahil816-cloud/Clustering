
# Clustering


## Question 1: Difference between K-Means and Hierarchical Clustering (with use cases)

K-Means is a partition-based algorithm that divides data into K clusters by minimizing distance to cluster centers.
Hierarchical clustering builds clusters in a tree-like structure without fixing cluster count in advance.

Use K-Means for customer segmentation and hierarchical clustering for gene or document grouping.



## Question 2: Purpose of Silhouette Score

Silhouette Score measures how well data points fit within their cluster compared to others.
It helps evaluate cluster quality and select optimal cluster numbers.



## Question 3: Core Parameters of DBSCAN

DBSCAN uses eps (neighborhood radius) and min_samples (points to form dense region).
These control cluster shape, density and noise detection.



## Question 4: Importance of Feature Scaling

Clustering depends on distances. Without scaling, large-value features dominate.
Scaling ensures fair contribution of all features.



## Question 5: Elbow Method

Elbow method plots K vs inertia and finds point where improvement slows.
This indicates optimal number of clusters.


## Question 6: K-Means on make_blobs Dataset

```python

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300, centers=4, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, c='red', marker='X')
plt.show()
```

## Question 7: DBSCAN on Wine Dataset

```python

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

X, y = load_wine(return_X_y=True)
X_scaled = StandardScaler().fit_transform(X)

db = DBSCAN(eps=1.5, min_samples=5)
labels = db.fit_predict(X_scaled)

n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print("Clusters found:", n_clusters)
```

## Question 8: DBSCAN on make_moons with Outliers

```python

from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

X, _ = make_moons(n_samples=200, noise=0.1, random_state=42)

db = DBSCAN(eps=0.3, min_samples=5)
labels = db.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels, cmap='plasma')
plt.show()
```

## Question 9: Agglomerative Clustering after PCA

```python

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

X, y = load_wine(return_X_y=True)
X_scaled = StandardScaler().fit_transform(X)

X_pca = PCA(n_components=2).fit_transform(X_scaled)

model = AgglomerativeClustering(n_clusters=3)
labels = model.fit_predict(X_pca)

plt.scatter(X_pca[:,0], X_pca[:,1], c=labels, cmap='viridis')
plt.show()
```
